# Sample Account Data

downloaded some sample data sets to populate CRM tables for MHG Bluesky project.

Data Source:  [Free Sample Data](http://www.briandunning.com)


```python
import numpy as np
import datetime as dt
import pandas as pd
import uuid
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost/postgres')
```

```python
df = pd.read_csv('au-500.csv', dtype=str)
df = df.rename({"post": "postal_code", "state": "state_or_province"}, axis="columns")
df['country'] = 'AU'
raw = pd.read_csv('ca-500.csv', dtype=str)
raw = raw.rename({"postal": "postal_code", "province": "state_or_province"}, axis="columns")
raw['country'] = 'CA'
df = pd.concat([df, raw])
raw = pd.read_csv('uk-500.csv', dtype=str)
raw = raw.rename({"postal": "postal_code", "county": "state_or_province"}, axis="columns")
raw['country'] = 'GB'
df = pd.concat([df, raw])
raw = pd.read_csv('us-500.csv', dtype=str)
raw = raw.rename({"zip": "postal_code", "state": "state_or_province"}, axis="columns")
raw['country'] = 'US'
df = pd.concat([df, raw])
df['addrId'] = pd.factorize(df.address+df.city+df.postal_code+df.state_or_province+df.country)[0]
df['acctId'] = pd.factorize(df.company_name)[0]
df['address_id'] = df.apply(lambda x: str(uuid.uuid4()), axis=1)
df['account_id'] = df.apply(lambda x: str(uuid.uuid4()), axis=1)
df['contact_id'] = df.apply(lambda x: str(uuid.uuid4()), axis=1)
df['created_timestamp'] = dt.datetime.now()
df['updated_timestamp'] = dt.datetime.now()
```

```python

df['acctId'] = pd.factorize(df.company_name)[0]
df
```

```python
addys = df[[
    'address_id',
    'created_timestamp',
    'updated_timestamp',
    'address',
    'city',
    'postal_code',
    'state_or_province',
    'country']].copy()
addys['address_type'] = 'O'
addys = addys.rename({"address": "address_line1", "address_id": 'id'}, axis="columns")
addys['postal_code'] = addys.apply(lambda x: str(x['postal_code'])[:10], axis=1)
addys['state_or_province'] = addys.apply(lambda x: str(x['state_or_province'])[:10], axis=1)
addys.to_sql('address', engine, chunksize=1000, if_exists='append', index=False)
```

```python
ind = pd.read_sql_table('industry_type', engine)
```

```python
def get_parent(acctId):
    matches = df[df['acctId'] == acctId]
    return matches[:1]['account_id'].to_string(index=False)

get_parent(766)
```

```python
accts = df[[
    'account_id',
    'created_timestamp',
    'updated_timestamp',
    'company_name',
    'phone1',
    'address_id',
    'acctId']].copy()
accts['description'] = accts.apply(lambda x: 'description for: %s' % x['company_name'], axis=1)
accts['ext_id'] = accts.apply(lambda x: 'ext_%s' % x['account_id'][:8], axis=1)
accts['industry'] = accts.apply(lambda x: ind['code'][x['acctId'] % 34], axis=1)
accts['owner_id'] = '2'
accts = accts.rename({
    "company_name": "account_name",
    "account_id": 'id',
    'phone1': 'phone',
    'address_id': 'billing_address'}, axis="columns")
accts['parent_id'] = accts.apply(lambda x: get_parent(x['acctId']), axis=1)
accts['parent_id'] = accts.apply(lambda x: None if x['parent_id'] == x['id'] else x['parent_id'], axis=1)
accts = accts.drop(['acctId'], axis=1)
accts
```

```python
accts[accts['parent_id'].isnull()].to_sql('account', engine, chunksize=1000, if_exists='append', index=False)
```

```python
accts[accts['parent_id'].notnull()].to_sql('account', engine, chunksize=1000, if_exists='append', index=False)
```
