# Reset Postgres dB

Removes all tables from local Postgres

```python
import pandas as pd
import uuid
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost/postgres')
```

```python
stmt_factory = """select 'drop table ' || tablename || ' cascade' as STMT
from pg_tables
where schemaname = 'public'"""
stmts = pd.read_sql(stmt_factory, engine)
stmts
```

```python
conn = engine.connect()
for ds in stmts['stmt'].tolist():
    print(conn.execute(ds))
```
